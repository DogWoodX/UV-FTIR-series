```python
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv('4h.csv')

# Melt the DataFrame to long format
df_melted = df.melt(id_vars=['Lambda'], var_name='Series', value_name='abs')

# Create the scatter plot
fig = px.scatter(df_melted, x='Lambda', y='abs', color='Series')

# Show the plot
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv('4h.csv')

# Melt the DataFrame to long format
df_melted = df.melt(id_vars=['Lambda'], var_name='Series', value_name='abs')

# Create the line plot
fig = px.line(df_melted, x='Lambda', y='abs', color='Series')

# Show the plot
fig.show()
```
