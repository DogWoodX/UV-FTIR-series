### ***1 hora 200 rpm 50 Â°C***

```python
import pandas as pd
import plotly.express as px
import os

# Define the file path for different operating systems
file_path = r'D:/OneDog/OneDrive/IrohLabs/JOE/JoeVV/df/1hCALBconcentrada.csv'

# Adjust file path for Unix-like systems (Linux, macOS) if necessary
if os.name != 'nt':  # 'nt' is the name for Windows systems
    file_path = file_path.replace('D:', '/mnt/d').replace('\\', '/')

# Load the data
df = pd.read_csv(file_path)

# Melt the DataFrame to long format
df_melted = df.melt(id_vars=['Lambda'], var_name='Series', value_name='abs')

# Create the line plot
fig = px.line(df_melted, x='Lambda', y='abs', color='Series')

# Show the plot
fig.show()
```
